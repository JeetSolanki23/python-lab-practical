```python
import pandas as pd

# Define two Pandas Series
series1 = pd.Series([10, 20, 30, 40, 50])
series2 = pd.Series([1, 2, 3, 4, 5])

# Perform arithmetic operations
addition_result = series1 + series2
subtraction_result = series1 - series2
multiplication_result = series1 * series2
division_result = series1 / series2

# Display the results
print("Addition Result:")
print(addition_result)

print("\nSubtraction Result:")
print(subtraction_result)

print("\nMultiplication Result:")
print(multiplication_result)

print("\nDivision Result:")
print(division_result)
```

```python
import matplotlib.pyplot as plt

# Data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

# Vertical Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(languages, popularity, color=colors)

# Adding text labels above each bar
for i in range(len(languages)):
    plt.text(i, popularity[i] + 0.5, str(popularity[i]), ha='center', va='bottom')

plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Languages (Vertical Bar Chart)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Horizontal Bar Chart
plt.figure(figsize=(10, 6))
plt.barh(languages, popularity, color=colors)

# Adding text labels above each bar
for i in range(len(languages)):
    plt.text(popularity[i] + 0.2, i, str(popularity[i]), ha='left', va='center')

plt.xlabel('Popularity')
plt.ylabel('Programming Languages')
plt.title('Popularity of Programming Languages (Horizontal Bar Chart)')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
countries = ['India', 'United States', 'Great Britain', 'China', 'Russia', 'Germany']
gold_medals = [50, 46, 27, 26, 19, 17]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(gold_medals, labels=countries, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# Add a title
plt.title('Gold Medal Achievements of Five Most Successful Countries in 2022 Summer Olympics')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the plot
plt.show()
```
